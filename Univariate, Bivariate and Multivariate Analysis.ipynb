```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("H:\Krish Naik\Iris Dataset\Iris Dataset.csv")
```

### Univariate Analysis

```python
df_setosa=df[df["species"]=="Setosa"]
```

```python
df_virginica=df.loc[df["species"]=="Virginica"]
df_versicolor=df.loc[df["species"]=="Versicolor"]
```

```python
plt.plot(df_setosa["sepal_length"])
```

```python
plt.plot(df_virginica["sepal_length"], np.zeros_like(df_virginica["sepal_length"]),'o')
plt.show()
# since it is an univariate variable so we created y-axis as zero
# after zero_like("how many number of zeros you want")
```

```python
plt.plot(df_setosa["sepal_length"], np.zeros_like(df_setosa["sepal_length"]),'o')
plt.show()
```

```python
plt.plot(df_versicolor["sepal_length"], np.zeros_like(df_versicolor["sepal_length"]),'o')
plt.show()
```

```python
plt.plot(df_setosa["sepal_length"], np.zeros_like(df_setosa["sepal_length"]),'o')
plt.plot(df_virginica["sepal_length"], np.zeros_like(df_virginica["sepal_length"]),'o')
plt.plot(df_versicolor["sepal_length"], np.zeros_like(df_versicolor["sepal_length"]),'o')
plt.xlabel("Sepal Length")
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue="species",height=5).map(plt.scatter,"sepal_length", "sepal_width").add_legend()
#hue= what are the feature we want to categories with
plt.show()
```

```python
sns.FacetGrid(df,hue="species",height=5).map(plt.scatter,"petal_length", "sepal_width").add_legend()
plt.show()
```

### Multivariate Analysis

```python
sns.pairplot(df,hue="species", height=3)
plt.show()
```
